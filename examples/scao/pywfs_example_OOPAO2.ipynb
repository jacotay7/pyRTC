# pyRTC Example : Simulating a Single Conjugate Adaptive Optics (SCAO) System with a Pyramid Wavefront Sensor (PYWFS)

### This tutorial walks you through how to build and run a basic SCAO system using pyRTC, with OOPAO providing the simulated telescope, deformable mirror (DM), and atmosphere.

You’ll learn how to:

* Load and interpret a YAML configuration file

* Initialize each AO subsystem (WFS, DM, PSF, etc.)

* Run the control loop and visualize results

* Compute an interaction matrix (IM) and evaluate PSFs

## 1. Overview

A Single Conjugate Adaptive Optics (SCAO) system uses one wavefront sensor and one deformable mirror conjugated to a single atmospheric layer to correct wavefront distortions.

In this example:

* The Pyramid Wavefront Sensor (PYWFS) measures the incoming wavefront.

* The Deformable Mirror (DM) applies the correction.

* The Science Camera (PSF) measures the corrected image quality.

* pyRTC coordinates the data flow in real-time between components.

### ⚠️ Important Before You Start

This notebook runs all components in a single Python process.
That makes it easier to debug but slower because of the Python Global Interpreter Lock (GIL).

For true real-time performance, use the Hardware Launcher tool — see
pywfs_example_OOPAO_pipeline.py for details.

## 2. Import Dependencies

```python
import matplotlib.pyplot as plt
import time

# Import pyRTC Core classes and OOPAO interface
from pyRTC import *
from pyRTC.hardware.OOPAOInterface import OOPAOInterface
```

## 4. Load Configuration File

All simulation parameters are defined in a YAML config file — this includes telescope, WFS, DM, PSF, and loop settings.

```python
conf = utils.read_yaml_file("pywfs_OOPAO_config.yaml")

# Split the config into component-specific sections
confLoop = conf["loop"]
confWFS = conf["wfs"]
confWFC = conf["wfc"]
confPSF = conf["psf"]
confSlopes = conf["slopes"]

print(confLoop)
print(confWFS)
print(confWFC)
print(confPSF)
print(confSlopes)
```

## 5. Initialize the AO Hardware Simulation

We use the OOPAO interface to create the simulated AO components.

This interface initializes:

* the wavefront sensor (wfs)

* the deformable mirror (dm)

* the science camera (psf)

* and the slopes processor

```python
# Create OOPAO simulation interface
sim = OOPAOInterface(conf=conf, param=None)

# Extract references to the simulated hardware
wfs, dm, psf = sim.get_hardware()
```

## 6. Define the Modal Basis (KL Modes)

Before running the loop, we define a modal basis for the DM.
Here, we use the Karhunen-Loève (KL) basis, which represents statistically optimal modes for atmospheric turbulence.

```python
from OOPAO.calibration.compute_KL_modal_basis import compute_KL_basis

NUM_MODES = confWFC["numModes"]  # must be less than total KL modes

M2C_KL = compute_KL_basis(sim.tel, sim.atm, sim.dm)
dm.setM2C(M2C_KL[:, :NUM_MODES])
```

## 7. Initialize the Slopes Processor and Control Loop

The SlopesProcess converts WFS signals into slope measurements,
and the Loop object controls the feedback from WFS → DM.

```python
slopes = SlopesProcess(conf=confSlopes)
loop = Loop(conf=confLoop)
```

## 8. Start AO Components

Now start each process. This activates real-time computations as defined in the config file.

```python
dm.start()
dm.flatten()  # Start from a flat mirror

wfs.start()
slopes.start()

print(sim.dm.OPD.shape)  # Check DM surface shape

psf.start()
dm.flatten()
slopes.plotPupils()  # Visualize pupil masks
```

## 9. Compute the Interaction Matrix (IM)

The Interaction Matrix (IM) characterizes how each DM actuator affects the WFS signal — essential for building the control matrix.

```python
# Temporarily remove atmosphere to measure IM in a static environment
sim.removeAtmosphere()

psf.takeModelPSF()  # Reference PSF for Strehl calculation

loop.pokeAmp = 1e-7  # Small actuator perturbation amplitude
loop.computeIM()      # Compute IM (blocking call)

# Restore atmosphere
sim.addAtmosphere()

# Visualize
plt.imshow(sim.dm.OPD)
loop.plotIM()
plt.show()
```

## 10. Run the AO Loop

Now run the adaptive optics feedback loop.

```python
dm.flatten()
time.sleep(1e-2)

loop.setGain(0.3)  # Loop gain controls correction strength
loop.start()

time.sleep(10)  # Run for 10 seconds

loop.stop()
dm.flatten()
```

## 11. Visualize Results

Plot several outputs from the simulation.

```python
# Science camera PSF (raw)
plt.imshow(psf.data)
plt.colorbar()
plt.title("Science Camera PSF (Current Frame)")
plt.show()

# Normalized PSF from OOPAO
plt.imshow(psf.tel.PSF_norma)
plt.colorbar()
plt.title("Normalized PSF (OOPAO)")
plt.show()

# Final DM optical path difference (OPD)
plt.imshow(sim.dm.OPD)
plt.colorbar()
plt.title("Final DM Shape (OPD Map)")
plt.show()

# WFS 2D signal
plt.imshow(sim.wfs.signal_2D)
plt.colorbar()
plt.title("WFS 2D Signal")
plt.show()

# Processed slopes output
plt.imshow(slopes.signal2D.read_noblock())
plt.colorbar()
plt.title("Processed Slopes (2D)")
plt.show()
```
